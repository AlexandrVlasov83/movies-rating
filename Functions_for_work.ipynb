### Functions

```python
import pandas as pd
```

```python
# импорт файлов для работы
M = pd.read_csv('D:/Нетология/Проекты/Movies/source/movies.csv')
R = pd.read_csv('D:/Нетология/Проекты/Movies/source/ratings.csv')
```

```python
# объединяем две таблицы для работы
movies_rating = M.merge(R, on='movieId', how='left')
```

```python
movies_rating
```

```python
# группируем данные и получаем средний рейтинг
movies_rating_mean = movies_rating.groupby(['movieId','title','genres']).rating.mean().to_frame().reset_index()
```

```python
movies_rating_mean
```

```python
def movie_class(data):
    """
    Функция для определения уровня рейтинга для фильма из списка
    """
    if data <=2:
        return 'low'
    elif 2 < data <= 4:
        return 'middle'
    return 'high'
```

```python
# добавляем в таблицу колонку с уровнем рейтинга
movies_rating_mean['level'] = movies_rating_mean.rating.apply(movie_class)
```

```python
movies_rating_mean
```

```python
years = list(range(1900,2023))
```

```python
def production_year(data):
    """
    Функция для выделения из текста года производства фильма
    """
    for i in years:
        if str(i) in data:
            return int(i)
    return 'Null'
```

```python
# добавление колонки с годом производства
movies_rating_mean['production_year'] = movies_rating_mean.title.apply(production_year)
movies_result = movies_rating_mean
```

```python
movies_result
```

```python
def movie_old(data):
    """
    Функция для распределения фильмов по старости выхода в прокат
    """
    if 2023 > data >= 2017:
        return 'new'
    elif 2000 <= data < 2017:
        return '20th century'
    elif 1960 <= data < 2000:
        return 'old'
    elif 1900 <= data < 1960:
        return 'retro'
    return 'Null'
```

```python
# добавление в таблицу колонки с уровнем старости выхода в прокат фильма
movies_result['old'] = movies_result.production_year.apply(movie_old)
```

```python
movies_result
```

```python
movies_result.to_csv('movies_result')
```
